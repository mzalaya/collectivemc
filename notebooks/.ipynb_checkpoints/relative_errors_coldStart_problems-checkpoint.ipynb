```python
# NUMPY
import numpy as np

# PANDAS
import pandas as pd

# MATPLOTLIB
import matplotlib.pyplot as plt
%matplotlib inline 
from matplotlib import rc
rc('font', **{'family': 'sans-serif', 'sans-serif': ['Computer Modern Roman']})
params = {'axes.labelsize': 8, # 12
          'font.size': 8, # 12
          'legend.fontsize': 8, # 12
          'xtick.labelsize': 8, # 10
          'ytick.labelsize': 8, # 10
          'text.usetex': True,
          'figure.figsize': (16, 6)}
plt.rcParams.update(params)

# SEABORN 
import seaborn as sns
sns.set_context("poster")
sns.set_style("ticks")
```

```python
#
relative_erros_collect1 = [r'$RE(\widehat{\mathcal{M}}^1_{cold}, {\mathcal{M}}^1_{cold})$']*10
relative_errors_gaussian = [r'$RE(\widehat{{M}}^1_{cold}, {{M}}^1_{cold})$']*10
relative_erros_collect2 = [r'$RE(\widehat{\mathcal{M}}^2_{cold}, {\mathcal{M}}^2_{cold})$']*10
relative_errors_poisson = [r'$RE(\widehat{{M}}^2_{cold}, {{M}}^2_{cold})$']*10
relative_erros_collect3 = [r'$RE(\widehat{\mathcal{M}}^3_{cold}, {\mathcal{M}}^3_{cold})$']*10
relative_errors_bernoulli = [r'$RE(\widehat{{M}}^3_{cold}, {{M}}^3_{cold})$']*10

relatives_errors_letters = [relative_erros_collect1,
                   relative_errors_gaussian,
                   relative_erros_collect2,
                   relative_errors_poisson,
                   relative_erros_collect3,
                   relative_errors_bernoulli]
```

```python
relatives_errors_letters  = [item for sublist in relatives_errors_letters for item in sublist]
```

```python
relative_errors_3000 = relative_errors_3000.values
relative_errors_6000 = relative_errors_6000.values
relative_errors_9000 = relative_errors_9000.values
```

```python
relative_errors_3000 = relative_errors_3000.T.ravel()
relative_errors_6000 = relative_errors_6000.T.ravel()
relative_errors_9000 = relative_errors_9000.T.ravel()
```

```python
relative_errors_9000
```

```python
data_relative_errors = pd.DataFrame({
    'Relative Errors': relatives_errors_letters,
    r'$d = 3000$': relative_errors_3000,
    r'$d = 6000$': relative_errors_6000,
    r'$d = 9000$': relative_errors_9000
})
    
columns = ['Relative Errors', r'$d = 3000$', r'$d = 6000$', r'$d = 9000$']
data_relative_errors = data_relative_errors[columns]
```

```python
import matplotlib.pyplot as plt

data = data_relative_errors.set_index('Relative Errors').unstack().reset_index()
data.columns = ['Dimensions','Relative Errors', 'Relative errors']

colors = plt.cm.Set1(np.linspace(0, .5, 6))
palette = {r'$RE(\widehat{\mathcal{M}}^1_{cold}, {\mathcal{M}}^1_{cold})$': colors[2],
     r'$RE(\widehat{{M}}^1_{cold}, {{M}}^1_{cold})$': colors[0],
     r'$RE(\widehat{\mathcal{M}}^2_{cold}, {\mathcal{M}}^2_{cold})$': colors[3],
     r'$RE(\widehat{{M}}^2_{cold}, {{M}}^2_{cold})$': colors[4],
     r'$RE(\widehat{\mathcal{M}}^3_{cold}, {\mathcal{M}}^3_{cold})$': colors[1],
     r'$RE(\widehat{{M}}^3_{cold}, {{M}}^3_{cold})$': colors[5]
    }

fig = plt.figure(figsize=(16, 8))
sns.set(style="white")
g = sns.factorplot(x='Dimensions',
                   y= 'Relative errors',
                   hue='Relative Errors',
                   data=data,
                   #kind='box',
                   kind='bar', 
                   #kind="swarm",
                   aspect=2,
                   palette=palette,
                   legend=False)
g.set_ylabels(fontsize = 15)
g.set_xlabels(fontsize = 15)
# g.set_xticklabels(rotation=50, fontsize = 13)
g.set_xticklabels(fontsize = 15)
g.add_legend(fontsize = 13)
g.set_yticklabels(fontsize = 15)
g.fig.get_axes()[0].set_yscale('symlog')
#g.map(pl.axvline(x=0.03, ymin=0.3, ymax=1.0))#, color='k', linestyle='--', alpha=0.3))
plt.savefig('./figs/relative_errors_ColdStarProblems.pdf', bbox_inches='tight')
plt.show();
```
