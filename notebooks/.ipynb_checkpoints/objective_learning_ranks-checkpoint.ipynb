```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(26, 16))
colors = plt.cm.Set1(np.linspace(0, .5, 4))
# First row
axes[0,0].set_yscale('symlog')
axes[0,0].plot(times_3000, objectives_3000, color=colors[0], 
               linewidth=5, label='Collective' ,marker='o', markersize=14)
axes[0,0].plot(times_3000_gaussian, objectives_3000_gaussian, 
               color=colors[1], linewidth=5, label='Gaussian',marker='>', markersize=14)
axes[0,0].plot(times_3000_poisson, objectives_3000_poisson, 
               color=colors[2], linewidth=5, label='Poisson' ,marker='D', markersize=12)
axes[0,0].plot(times_3000_bernoulli, objectives_3000_bernoulli, 
               color=colors[3], linewidth=5, label='Bernoulli' ,marker='X', markersize=14)
axes[0,0].legend();
axes[0,0].set_title(r'$d=3000$');
axes[0,0].set_xlabel('Times (seconds)')
axes[0,0].set_ylabel(r'$\mathcal{F}_\lambda$')

#
axes[0,1].set_yscale('symlog')
axes[0,1].plot(times_6000, objectives_6000, 
               color=colors[0], linewidth=5, label='Collective' ,marker='o', markersize=14)
axes[0,1].plot(times_6000_gaussian, objectives_6000_gaussian, 
               color=colors[1], linewidth=5, label='Gaussian' ,marker='>', markersize=14)
axes[0,1].plot(times_6000_poisson, objectives_6000_poisson, 
               color=colors[2], linewidth=5, label='Poisson' ,marker='D', markersize=12)
axes[0,1].plot(times_6000_bernoulli, objectives_6000_bernoulli, 
               color=colors[3], linewidth=5, label='Bernoulli',marker='X', markersize=14)
axes[0,1].legend();
axes[0,1].set_title(r'$d=6000$');
axes[0,1].set_xlabel('Times (seconds)')

# 
axes[0,2].set_yscale('symlog')
axes[0,2].plot(times_9000, objectives_9000, 
               color=colors[0], linewidth=5, label='Collective' ,marker='o', markersize=14)
axes[0,2].plot(times_9000_gaussian, objectives_9000_gaussian, 
               color=colors[1], linewidth=5, label='Gaussian' ,marker='>', markersize=14)
axes[0,2].plot(times_9000_poisson, objectives_9000_poisson, 
               color=colors[2], linewidth=5, label='Poisson' ,marker='D', markersize=12)
axes[0,2].plot(times_9000_bernoulli, objectives_9000_bernoulli, 
               color=colors[3], linewidth=5, label='Bernoulli' ,marker='X', markersize=14)
axes[0,2].legend();
axes[0,2].set_title(r'$d=9000$');
axes[0,2].set_xlabel('Times (seconds)')

# Second row
axes[1,0].set_yscale('symlog')
axes[1,0].plot(-np.log(lambdas_3000), objectives_3000, 
               color=colors[0], linewidth=5, label='Collective' ,marker="o", markersize=14)
axes[1,0].plot(-np.log(lambdas_3000_gaussian), objectives_3000_gaussian, 
               color=colors[1], linewidth=5, label='Gaussian' ,marker='>', markersize=14)
axes[1,0].plot(-np.log(lambdas_3000_poisson), objectives_3000_poisson, 
               color=colors[2], linewidth=5, label='Poisson' ,marker='D', markersize=12)
axes[1,0].plot(-np.log(lambdas_3000_bernoulli), objectives_3000_bernoulli, 
               color=colors[3], linewidth=5, label='Bernoulli' ,marker='X', markersize=14)
axes[1,0].legend();
axes[1,0].set_title(r'$d=3000$');
axes[1,0].set_xlabel(r'$-\log(\lambda)$');
axes[1,0].set_ylabel(r'$\mathcal{F}_\lambda$')


#
axes[1,1].set_yscale('symlog')
axes[1,1].plot(-np.log(lambdas_6000), (objectives_6000), 
               color=colors[0], linewidth=5, label='Collective' ,marker='o', markersize=14)
axes[1,1].plot(-np.log(lambdas_6000_gaussian), objectives_6000_gaussian, 
               color=colors[1], linewidth=5, label='Gaussian' ,marker='>', markersize=14)
axes[1,1].plot(-np.log(lambdas_6000_poisson), objectives_6000_poisson, 
               color=colors[2], linewidth=5, label='Poisson' ,marker='D', markersize=12)
axes[1,1].plot(-np.log(lambdas_6000_bernoulli), objectives_6000_bernoulli, 
               color=colors[3], linewidth=5, label='Bernoulli' ,marker='X', markersize=14)
axes[1,1].legend();
axes[1,1].set_title(r'$d=6000$');
axes[1,1].set_xlabel(r'$-\log(\lambda)$');

# 
axes[1,2].set_yscale('symlog')
axes[1,2].plot(-np.log(lambdas_9000), (objectives_9000), 
               color=colors[0], linewidth=5, label='Collective' ,marker='o', markersize=14)
axes[1,2].plot(-np.log(lambdas_9000_gaussian), objectives_9000_gaussian, 
               color=colors[1], linewidth=5, label='Gaussian' ,marker='>', markersize=14)
axes[1,2].plot(-np.log(lambdas_9000_poisson), objectives_9000_poisson, 
               color=colors[2], linewidth=5, label='Poisson' ,marker='D', markersize=12)
axes[1,2].plot(-np.log(lambdas_9000_bernoulli), objectives_9000_bernoulli, 
               color=colors[3], linewidth=5, label='Bernoulli' ,marker='X', markersize=14)
axes[1,2].legend();
axes[1,2].set_xlabel(r'$-\log(\lambda)$');
axes[1,2].set_title(r'$d=9000$');
plt.subplots_adjust(hspace=.5)
plt.tight_layout()
fig.savefig('./figs/objectives.pdf', bbox_inches='tight')
```

```python
fig, axes= plt.subplots(ncols=3, figsize=(26, 8))
colors = plt.cm.Set1(np.linspace(0, .5, 4))

# 
axes[0].plot(range((len(times_3000))-8), ranksin_3000[:-8], 
              color='g', linewidth=5, label='RanksIn' ,marker=">", markersize=14)
axes[0].plot(range(len(times_3000)-8), ranksout_3000[:-8], 
              color='c', linewidth=5, label='RanksOut' ,marker="<", markersize=14)
axes[0].legend()
axes[0].set_title(r'$d=3000$')
axes[0].set_xlabel('Iterations');
axes[0].set_ylabel('Learned ranks');
axes[0].set_ylim(ymin=3);
# 
axes[1].plot(range(len(times_6000)-6), ranksin_6000[:-6], 
             color='g', linewidth=5, label='RanksIn' ,marker=">", markersize=14)
axes[1].plot(range(len(times_6000)-6), ranksout_6000[:-6],
             color='c', linewidth=5, label='RanksOut' ,marker="<", markersize=14)
axes[1].set_title(r'$d=6000$')
axes[1].set_xlabel('Iterations');
axes[1].set_ylabel('Learned ranks');
axes[1].legend()
axes[1].set_ylim(ymin=3);
# 
axes[2].plot(range(len(times_9000)), ranksin_9000, 
             color='g', linewidth=5, label='RanksIn' ,marker=">", markersize=14)
axes[2].plot(range(len(times_9000)), ranksout_9000,
             color='c', linewidth=5, label='RanksOut' ,marker="<", markersize=14)
axes[2].legend()
axes[2].set_title(r'$d=9000$');
axes[2].set_ylim(ymin=3);
axes[2].set_ylabel('Learned ranks');
axes[2].set_xlabel('Iterations');
plt.tight_layout()
fig.savefig('./figs/learning_ranks.pdf', bbox_inches='tight')
```
