```python
# NUMPY
import numpy as np

# PANDAS
import pandas as pd

# MATPLOTLIB
import matplotlib.pyplot as plt
%matplotlib inline 
#from matplotlib import rc
#rc('font', **{'family': 'sans-serif', 'sans-serif': ['Computer Modern Roman']})
params = {'axes.labelsize': 8, # 12
          'font.size': 8, # 12
          'legend.fontsize': 8, # 12
          'xtick.labelsize': 8, # 10
          'ytick.labelsize': 8, # 10
          'text.usetex': True,
          'figure.figsize': (10, 8)}
plt.rcParams.update(params)

# SEABORN 
import seaborn as sns
sns.set_context("poster")
sns.set_style("ticks")
```

```python
relative_errors_3000 = pd.read_csv('./tables/RMSE_relatives_3000.csv')
relative_errors_6000 = pd.read_csv('./tables/RMSE_relatives_6000.csv')
relative_errors_9000 = pd.read_csv('./tables/RMSE_relatives_9000.csv')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(26, 8))
colors = plt.cm.Set1(np.linspace(0, .5, 4))
# fig 1
ax1.plot(relative_errors_3000['probas'], relative_errors_3000['RMSE_relative_collect'], color=colors[0],
          linewidth=5, label='Collective' ,marker="o", markersize=14)
ax1.plot(relative_errors_3000['probas'], relative_errors_3000['RMSE_relative_gaussian'], color=colors[1],
          linewidth=5, label='Gaussian' ,marker='>', markersize=14)
ax1.plot(relative_errors_3000['probas'], relative_errors_3000['RMSE_relative_poisson'], color=colors[2],
         linewidth=5, label='Poisson' ,marker='D', markersize=12)
ax1.plot(relative_errors_3000['probas'], relative_errors_3000['RMSE_relative_bernoulli'], color=colors[3],
          linewidth=5, label='Bernoulli' ,marker='X', markersize=14)
ax1.set_xlim(0.05, 1)
ax1.set_xlabel('Percentage of known entries')
ax1.set_ylabel(r'Relative errors')
ax1.legend(loc="lower left") #(handles=[blue_line, reds_line]);
ax1.set_title('$d=3000$');

# fig 2
ax2.plot(relative_errors_6000['probas'], relative_errors_6000['RMSE_relative_collect'], color=colors[0],
          linewidth=5, label='Collective',marker="o", markersize=14)
ax2.plot(relative_errors_6000['probas'], relative_errors_6000['RMSE_relative_gaussian'], color=colors[1],
          linewidth=5, label='Gaussian' ,marker='>', markersize=14)
ax2.plot(relative_errors_6000['probas'], relative_errors_6000['RMSE_relative_poisson'], color=colors[2],
          linewidth=5, label='Poisson' ,marker='D', markersize=12)
ax2.plot(relative_errors_6000['probas'], relative_errors_6000['RMSE_relative_bernoulli'], color=colors[3],
          linewidth=5, label='Bernoulli' ,marker='X', markersize=14)
ax2.set_xlim(0.05, 1)
ax2.set_xlabel('Percentage of known entries')
ax2.set_ylabel(r'Relative errors')
ax2.legend() #(handles=[blue_line, reds_line]);
ax2.set_title('$d=6000$');

# fig 3
ax3.plot(relative_errors_9000['probas'], relative_errors_9000['RMSE_relative_collect'], color=colors[0],
          linewidth=5, label='Collective' ,marker="o", markersize=14)
ax3.plot(relative_errors_9000['probas'], relative_errors_9000['RMSE_relative_gaussian'], color=colors[1],
          linewidth=5, label='Gaussian' ,marker='>', markersize=14)
ax3.plot(relative_errors_9000['probas'], relative_errors_9000['RMSE_relative_poisson'], color=colors[2],
          linewidth=5, label='Poisson' ,marker='D', markersize=12)
ax3.plot(relative_errors_9000['probas'], relative_errors_9000['RMSE_relative_bernoulli'], color=colors[3],
          linewidth=5, label='Bernoulli' ,marker='X', markersize=14)
ax3.set_xlim(0.05, 1)
ax3.set_xlabel('Percentage of known entries') #, fontsize = 15)
ax3.set_ylabel(r'Relative errors')
ax3.legend(loc="lower left") #(handles=[blue_line, reds_line]);
ax3.set_title('$d=9000$');
plt.tight_layout()
fig.savefig('./figs/relative_errors.pdf', bbox_inches='tight')
```
